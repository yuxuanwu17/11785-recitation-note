## Install google drive extension

#### Installation of jupyterlab
run in shell:

**conda install jupyterlab==2.1.1**

jupyterlab > 3.0 doesn't support extension.

#### Installation of extension
run in shell as adminstrator:

**conda install -c krinsman jupyterlab-google-drive**

## Setup of correct credentials
refer to:

https://github.com/jupyterlab/jupyterlab-google-drive/blob/master/docs/setup.md

https://towardsdatascience.com/integrate-jupyterlab-with-google-drive-98d13e340c63

Mistakes:

1. Don't enable "realtime api"

2. Add test user using your personal email
