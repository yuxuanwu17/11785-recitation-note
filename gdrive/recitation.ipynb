## Variable storage in notebooks


```python
from scipy.io import loadmat
from time import time
```

```python
# IDE work process
start = time()
data = loadmat('data.mat')
print(data.keys())
end = time()
print("IDE:", (end - start) * 100000) # sometimes, data is so large that it will take much time to load
```

```python
# Notebook work process
data = loadmat('data.mat')
```

```python
start = time()
print(data.keys())
end = time()
print("Notebook:", (end - start) * 100000)
```

```python
# PAT ATTENTION
data = {'1':'yes', '2':'no'}
print(data.keys())
```
